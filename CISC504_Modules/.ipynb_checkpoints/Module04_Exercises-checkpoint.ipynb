When opening a file, you may wonder how Python handles <i>closing</i> the file. Typically Python will handle that kind of thing automatically for you during garbage collection but if a code prematurely ends, the file may be left open which can cause performance degradation. You can insert a <code>f.close()</code> statement into your code but that can cause problems as well. To prevent these types of problems, its best to avoid simply calling <code>open()</code>. Instead we will use the following pattern.

```python
with open("pg37431.txt") as f:
    print(f.read(5))
```

Using a <code>with open</code> statement instead of just <code>open()</code> Python will only have the file open for the duration of the <code> with</code> statement. Closing it once the operation is over. This way you can open the file, store the data in a variable, close the file, and then manipulate the data as needed safely. Once you are done and need the write the file off to the file system  again, you can open the file again, save it, and Python will close it auto-magically. In the example above, we are using <code>read()</code> with a parameter of 5, indicating that we only want the first 5 characters in the file. We can also ask Python to give us only the first line of a text file using the <code>readline()</code> function. It's best to only get when you need from a file to help decrease the domain you are working with. This helps keep complexity down, performance up, and confusion at a minium when looking at files in a code environment.

```python
with open("pg37431.txt") as f:
    print(f.readline())
```

<p>Using the <code>datetime</code> and <code>time</code> modules, we are going to add time stamps to the log.txt file using a looping pattern and another file function named <code>write(n)</code> where <code>n</code> is the data you wish to write to the file. After which we will close the file. You should then go check your log.txt file in your file system and see if it worked! <br/><br/>
    Dont worry if you don't understand what <b>%Y%m%d_%H:%M:%S - </b> means. It's just a way of formatting a date string. <code>strfttime</code> is short for string format time <b>str </b> STRING <b> f </b> FORMAT <b>time </b> TIME
    </p>

```python
from datetime import datetime
import time

for i in range(0,10):
    print(datetime.now().strftime('%Y%m%d_%H:%M:%S - '),i)
    f.write(datetime.now().strftime('%Y%m%d_%H:%M:%S - '));
    time.sleep(1)
    f.write(str(i));
    f.write("\n")
f.close()
```

<h3>Defensive Programming</h3>
<p>
    In Python, because we have a loose typing structure, it is important to let the program check the values that are being passed around are valid. We can do that with assertion checks where we <code>assert</code> that a value adheres to a certain condition. If it doesn't, then we throw an <b>assertion</b> error and halt the program. This may seem like overkill when you are the only programmer using the code, you know where your code is being used and how you are using it, but it is very rare that only one developer works on a project, both in the industry and academia.
</p>

```python
def avg(marks):
    assert len(marks) != 0
    return round(sum(marks)/len(marks), 2)
```

```python
sem1_marks = [62, 65, 75]
print("Average marks for semester 1:",avg(sem1_marks))
```

Now all of that is not something we expect from you. This is actually the exact code you can pull from the matplot website. 

```python
import numpy as np
import matplotlib.pyplot as plt
def heatmap(data, row_labels, col_labels, ax=None,
            cbar_kw={}, cbarlabel="", **kwargs):
    if not ax:
        ax = plt.gca()

    im = ax.imshow(data, **kwargs)
    cbar = ax.figure.colorbar(im, ax=ax, **cbar_kw)
    cbar.ax.set_ylabel(cbarlabel, rotation=-90, va="bottom")
    ax.set_xticks(np.arange(data.shape[1]))
    ax.set_yticks(np.arange(data.shape[0]))
    ax.set_xticklabels(col_labels)
    ax.set_yticklabels(row_labels)
    ax.tick_params(top=True, bottom=False, labeltop=True, labelbottom=False)
    plt.setp(ax.get_xticklabels(), rotation=-30, ha="right", rotation_mode="anchor")
    for edge, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xticks(np.arange(data.shape[1]+1)-.5, minor=True)
    ax.set_yticks(np.arange(data.shape[0]+1)-.5, minor=True)
    ax.grid(which="minor", color="w", linestyle='-', linewidth=3)
    ax.tick_params(which="minor", bottom=False, left=False)
    return im, cbar
```

```python
data = np.array([
    [30, 20, 10,],
    [10, 40, 15],
    [12, 10, 20]
])

im, cbar = heatmap(data, ['Class-1', 'Class-2', 'Class-3'], ['A', 'B', 'C'], cmap='YlGn', cbarlabel='Number of Students')
```

Remember how we saved the heatmap by returning it? Well now that we did that, we can recreate the heatmap and annoate it. Now we have a way of displaying the map in mutliple ways without recreating the code. A great convience!

Again this is code that we can pull from the Matplot website. We aren't going to go through this one line by one but know that it can be found online <a href="https://matplotlib.org/gallery/images_contours_and_fields/image_annotated_heatmap.html?highlight=annotate%20heatmap">here<a/>. 

```python
def annotate_heatmap(im, data=None, valfmt="{x:.2f}",
                     textcolors=["black", "white"],
                     threshold=None, **textkw):
    import matplotlib
    if not isinstance(data, (list, np.ndarray)):
        data = im.get_array()
    if threshold is not None:
        threshold = im.norm(threshold)
    else:
        threshold = im.norm(data.max())/2.
    kw = dict(horizontalalignment="center",
              verticalalignment="center")
    kw.update(textkw)
    if isinstance(valfmt, str):
        valfmt = matplotlib.ticker.StrMethodFormatter(valfmt)
    texts = []
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            kw.update(color=textcolors[im.norm(data[i, j]) > threshold])
            text = im.axes.text(j, i, valfmt(data[i, j], None), **kw)
            texts.append(text)

    return texts
```

```python
im, cbar = heatmap(data, ['Class-1', 'Class-2', 'Class-3'], ['A', 'B', 'C'], cmap='YlGn', cbarlabel='Number of Students')

texts = annotate_heatmap(im, valfmt="{x}")
```

Finall we get to some 3-D plots using the <code>mpl_toolkits</code>. The MPL toolkit gives as 3-D plotting capabilities for seaborn and pyplot. You can see we aren't actually using <code>Axes3D</code> anywhere but if you try to execute this code without it, you'll get an error. That's because you need to bring in the capabilites and then <code>plt.figure()</code> object has the ability to use the <code>projection='3d'</code> parameter on line 12. This is just a demo graph and has not actually contation, its more to showcase the power of the MPL toolkit when imported into your project. 

```python
# from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
# import seaborn as sns
X = np.linspace(0, 10, 50)
Y = np.linspace(0, 10, 50)
X, Y = np.meshgrid(X, Y)
Z = (np.sin(X))

# Setup axis
fig = plt.figure(figsize=(7,5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)

# Add title and axes labels
ax.set_title("Demo of 3D Plot", size=13)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
